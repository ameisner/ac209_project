```python
import json
import numpy as np
f = open('data/gamespot.json')
data = json.load(f)
```

```python
rev_list = np.array([g['summary'] for g in data])
scores = np.array([g['score'] for g in data]).astype(float)

# get rid of empty reviews

u, ind_u = np.unique(rev_list, return_index=True)
rev_list = rev_list[ind_u]
scores = scores[ind_u]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# this token_pattern gets rid of non-alphabetical characters
vectorizer_alph_only = CountVectorizer(min_df=0.001, token_pattern=ur'\b[a-zA-Z][a-zA-Z]+\b')
vectorizer_alph_only.fit(rev_list)
words_alph_only = np.array(vectorizer_alph_only.get_feature_names())

vectorizer = CountVectorizer(min_df=0.001)
vectorizer.fit(rev_list)
words = np.array(vectorizer.get_feature_names())
garbage = set(words).difference(set(words_alph_only))
for g in garbage:
    print g,
    
print '\n' + '*'*40 + '\nnumber of features after removing garbage : ', len(words_alph_only)
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(ur'\b[a-zA-Z][a-zA-Z]+\b')
```

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
rev_list_stemmed = []
for rev in rev_list:
    tokens = tokenizer.tokenize(rev)
    stemmed = ""
    for token in tokens:
        stemmed = stemmed + " " + stemmer.stem(token)
    rev_list_stemmed.append(stemmed)
```

```python
def print_raw_stemmed(ix):
    print "RAW"
    print rev_list[ix]
    print "STEMMED"
    print rev_list_stemmed[ix] + '\n'
```
